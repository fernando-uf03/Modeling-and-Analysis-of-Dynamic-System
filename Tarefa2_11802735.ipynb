```python
import matplotlib.pyplot as plt
import numpy as np
import control as ctl
from control.matlab import *
# mais informações em https://python-control.readthedocs.io/en/0.8.3/intro.html

Kr=1
wn=4.5
qsi1=0.0
qsi2=0.4
qsi3=0.7
qsi4=1.0

plant_tf1 = TransferFunction(Kr*wn**2, [1,2*qsi1*wn,wn**2])
plant_tf2= TransferFunction(Kr*wn**2, [1,2*qsi2*wn,wn**2])
plant_tf3 = TransferFunction(Kr*wn**2, [1,2*qsi3*wn,wn**2])
plant_tf4 = TransferFunction(Kr*wn**2, [1,2*qsi4*wn,wn**2])

poles_tf1 = ctl.pole(plant_tf1)
poles_tf2 = ctl.pole(plant_tf2)
poles_tf3 = ctl.pole(plant_tf3)
poles_tf4 = ctl.pole(plant_tf4)
print("Polos da Função Transferência:")
print(poles_tf1)
print(poles_tf2)
print(poles_tf3)
print(poles_tf4)

plt.figure()
plt.title("Bode plot")
ctl.bode(plant_tf1)
ctl.bode(plant_tf2)
ctl.bode(plant_tf3)
ctl.bode(plant_tf4)
plt.legend(["qsi1=0","qsi2=0.4","qsi3=0.7","qsi4=1.0"])
plt.show()
```

```python
t=np.linspace(0,10,1001)
plt.figure()
t1,y1=ctl.step_response(plant_tf1,T=t)
t2,y2=ctl.step_response(plant_tf2,T=t)
t3,y3=ctl.step_response(plant_tf3,T=t)
t4,y4=ctl.step_response(plant_tf4,T=t)

plt.plot(t1,y1,'b',label='qsi=0')
plt.plot(t2,y2,'k',label='qsi=0.4')
plt.plot(t3,y3,'y',label='qsi=0.7')
plt.plot(t4,y4,'r',label='qsi=1')
plt.legend(["qsi1=0","qsi2=0.4","qsi3=0.7","qsi4=1.0"])
plt.show()
```

Quanto ao impacto do amortecimento, é possível notar que, quando o fator de amortecimento é nulo, o ganho oscila constantemente de forma limitada. Em sequência, assim que o fator de amortecimento passa a aumentar de 0 até 1, o sobre-sinal diminui até não existir (zeta=1).

```python
Kr=1
qsi=0.5
wn1=1
wn2=5
wn3=10
wn4=20
plant_tf1 = TransferFunction(Kr*wn1**2, [1,2*qsi*wn1,wn1**2])
plant_tf2= TransferFunction(Kr*wn2**2, [1,2*qsi*wn2,wn2**2])
plant_tf3 = TransferFunction(Kr*wn3**2, [1,2*qsi*wn3,wn3**2])
plant_tf4 = TransferFunction(Kr*wn4**2, [1,2*qsi*wn4,wn4**2])
poles_tf1 = ctl.pole(plant_tf1)
poles_tf2 = ctl.pole(plant_tf2)
poles_tf3 = ctl.pole(plant_tf3)
poles_tf4 = ctl.pole(plant_tf4)
print("Polos da Função Transferência:")
print(poles_tf1)
print(poles_tf2)
print(poles_tf3)
print(poles_tf4)
plt.figure()
plt.title("Bode plot")
ctl.bode(plant_tf1)
ctl.bode(plant_tf2)
ctl.bode(plant_tf3)
ctl.bode(plant_tf4)
plt.legend(["wn1=1","wn2=5","wn3=10","wn4=20"])
plt.show()
```

```python
t=np.linspace(0,10,1001)
plt.figure()
t1,y1=ctl.step_response(plant_tf1,T=t)
t2,y2=ctl.step_response(plant_tf2,T=t)
t3,y3=ctl.step_response(plant_tf3,T=t)
t4,y4=ctl.step_response(plant_tf4,T=t)

plt.plot(t1,y1,'b',label='wn=1')
plt.plot(t2,y2,'k',label='wn=5')
plt.plot(t3,y3,'y',label='wn=10')
plt.plot(t4,y4,'r',label='wn=20')
plt.legend(["wn1=1","wn2=5","wn3=10","wn4=20"])
plt.show()
```

Quanto ao impacto do aumento da frequência natural não amortecida, é possível notar que o aumento do valor de Wn, aumenta a velocidade de resposta dos sistema. Quanto maior Wn, maior a velocidade de resposta, ou seja, o sistema atinge a estabilidade em menos tempo.
