No primeiro caso, em que a frequência da entrada é menor do que Wn=sqrt(k/m), os gráficos de saída e de entrada estão em fase e a amplitude da saída é menor do que a amplitude da entrada. 

No segundo caso, em que a frequência da entrada é próxima a Wn=sqrt(k/m), o gráfico de saída está defasado em 90° com relação ao gráfico de entrada e a amplitude da saída, apesar de aumentar, é menor do que a amplitude da entrada.

Por fim, o terceiro caso, em que a frequência da entrada é maior do que Wn=sqrt(k/m), o gráfico de saída está defasado em 180° com relação ao gráfico de entrada, a amplitude da saída diminui com o tempo e é menor do que a amplitude da entrada.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import lsim
import control as ctl
from control import (TransferFunction)

m = 1
b = 0.9
k = 100

y0 = [00.0, 00.0]
t = np.linspace(0, 10, 1001)
u1 = np.sin((np.sqrt(k/m)-4)*t)
u2 = np.sin((np.sqrt(k/m))*t)
u3 = np.sin((np.sqrt(k/m)+10)*t)

plant_tf = TransferFunction(1, [m,b,k])
plant_tf

tout, yout = ctl.forced_response(plant_tf, U=u1, T=t)
plt.plot(tout, yout, 'b', linewidth=1.5, label='Saída')
#dividiu-se u1 por 4 para mudar a escala e facilitar a visualização
#sabe-se que na verdade a amplitude máxima da entrada é 1, em módulo
plt.plot(tout, u1/4, 'k', linewidth=1.5, label='Entrada') 
plt.legend(loc='best', shadow=True, framealpha=1)
plt.title('Gráfico 1')
plt.grid(alpha=0.3)
plt.xlabel('t')
plt.show()

tout, yout = ctl.forced_response(plant_tf, U=u2, T=t)
plt.plot(tout, yout, 'b', linewidth=1.5, label='Saída')
#dividiu-se u2 por 4 para mudar a escala e facilitar a visualização
#sabe-se que na verdade a amplitude máxima da entrada é 1, em módulo
plt.plot(tout, u2/4, 'k', linewidth=1.5, label='Entrada') 
plt.legend(loc='best', shadow=True, framealpha=1) 
plt.title('Gráfico 2')
plt.grid(alpha=0.3)
plt.xlabel('t')
plt.show()

tout, yout = ctl.forced_response(plant_tf, U=u3, T=t)
plt.plot(tout, yout, 'b', linewidth=1.5, label='Saída')
#dividiu-se u3 por 50 para mudar a escala e facilitar a visualização
#sabe-se que na verdade a amplitude máxima da entrada é 1, em módulo
plt.plot(tout, u3/50, 'k', linewidth=1.5, label='Entrada') 
plt.legend(loc='best', shadow=True, framealpha=1)
plt.title('Gráfico 3')
plt.grid(alpha=0.3)
plt.xlabel('t')
plt.show()
```
